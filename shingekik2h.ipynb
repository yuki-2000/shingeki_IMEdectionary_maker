```python
from google.colab import files
uploaded_files = files.upload()
uploaded_file_names = list(uploaded_files.keys())
uploaded_file_names.sort()
print(uploaded_file_names)
```

```python
from pykakasi import kakasi

kakasi = kakasi()
kakasi.setMode('J', 'H') #漢字からひらがなに変換
kakasi.setMode("K", "H") #カタカナからひらがなに変換
conv = kakasi.getConverter()

str = '東京タワー'

print(conv.do(str))


import pandas as pd
df = pd.read_excel(uploaded_file_names[0],header=None)

lst = list(df[0])
for i, s in enumerate(lst):
  print(df[0][i])
  print(conv.do(s))
  df[1][i]=conv.do(s)
df = df.iloc[:, [1,0,2]]

print(df)

```

```python
from google.colab import files
filename =  "shingekiIME.txt"
#df.to_csv(filename,  header=False, index=False, encoding = 'utf-8-sig', sep='\t') 
df.to_csv(filename,  header=False, index=False, encoding = 'UTF-16LE', sep='\t') 

files.download(filename)
```
